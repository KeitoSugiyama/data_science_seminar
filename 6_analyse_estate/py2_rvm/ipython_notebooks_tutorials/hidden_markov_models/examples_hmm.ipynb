# Bayesian Hidden Markov Models

```python
# required imports 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
% matplotlib inline
```

## Poisson Hidden Markov Model: Sunspot Data

```python
from skbayes.hidden_markov_models import VBPoissonHMM
# !!! don't forget data to the same directory that you use for running
# ipython notebook
sunspots = pd.read_csv('sunspot_data.csv')
X = sunspots['sunspot.month']
X = np.reshape(X,(X.shape[0],1))
dates = np.linspace(0,X.shape[0],X.shape[0])

# fit hmm with poisson emissions
model = VBPoissonHMM(n_hidden=4, n_init = 2, tol = 1e-3).fit(X)

# Predict the optimal sequence of hidden states
hidden_states = model.predict(X)
```

```python
# plot graphs
colors = ['ko','ro','bo','go']
plt.figure(figsize = (50,24))
for i in range(model.n_hidden):
    mask = hidden_states ==i
    im = plt.plot(dates[mask],X[mask,0],colors[i], markersize = 16, 
                  label = ("hidden state {0}".format(i)))
plt.legend()
plt.show()

# plot states by filled 
colors = ['black','red','blue','green']
fig, ax = plt.subplots(figsize = (50,24))
ax.plot(range(20))
for j in range(1,X.shape[0]):
    clr = colors[hidden_states[j]]
    ax.axvspan(dates[j-1],dates[j], color = clr)
plt.show()
```

## Gaussian Hidden Markov Model: Modelling returns and volume of trade for Apple stock

```python
# this example is similar to example in hmmlearn
# (code for importing data is taken from hmmlearn example )

import datetime
from skbayes.hidden_markov_models import VBGaussianHMM
from matplotlib import cm, pyplot as plt
from matplotlib.dates import YearLocator, MonthLocator
try:
    from matplotlib.finance import quotes_historical_yahoo_ochl
except ImportError:
    # For Matplotlib prior to 1.5.
    from matplotlib.finance import (
        quotes_historical_yahoo as quotes_historical_yahoo_ochl
    )

# Get quotes from Yahoo! finance
quotes = quotes_historical_yahoo_ochl(
    "AAPL", datetime.date(1995, 1, 1), datetime.date(2015, 1, 6))

# Unpack quotes
dates = np.array([q[0] for q in quotes], dtype=int)
close_v = np.array([q[2] for q in quotes])
volume = np.array([q[5] for q in quotes])[1:]

# compute returns
diff = np.diff(close_v)#/close_v[:-1]
dates = dates[1:]
close_v = close_v[1:]
# use returns & trading volume as training data
X = np.column_stack([diff, volume])
# fit Bayesian HMM
model = VBGaussianHMM(n_hidden=3, tol = 1e-3, n_iter=200, n_init = 5,
                      alpha_trans = 5, alpha_start = 1).fit(X)

# Predict the optimal sequence of hidden states using Viterbi
hidden_states = model.predict(X)
```

```python
fig, axs = plt.subplots(model.n_hidden, sharex=True, sharey=True,
                        figsize = (50,24))
colours = ['red','blue','green','yellow']
time_ax = np.linspace(0,X.shape[0]+1,X.shape[0]+1)
for i, (ax, colour) in enumerate(zip(axs, colours)):
    # Use fancy indexing to plot data in each state.
    mask = hidden_states == i
    idx = np.where(hidden_states==i)[0]
    ax.plot(time_ax[1:][mask], volume[mask],'ko', linewidth= 10)
    for id in idx:
        ax.axvspan(time_ax[id],time_ax[id+1], color = colour)
    ax.set_title("{0}th hidden state".format(i))
plt.show()


# plot states
fig, ax = plt.subplots(figsize = (50,24))
ax.plot(range(20))
for j in range(1,X.shape[0]):
    clr = colours[hidden_states[j]]
    ax.axvspan(time_ax[j-1],time_ax[j], color = clr)
plt.title("Most Likely States")
plt.show()
```

```python
fig, axs = plt.subplots(model.n_hidden, sharex=True, sharey=True,
                        figsize = (50,24))
for i, (ax, colour) in enumerate(zip(axs, colours)):
    # Use fancy indexing to plot data in each state.
    mask = hidden_states == i
    idx = np.where(hidden_states==i)[0]
    ax.plot(time_ax[1:][mask], close_v[mask],'ko', linewidth= 10)
    for id in idx:
        ax.axvspan(time_ax[id],time_ax[id+1], color = colour)
    ax.set_title("{0}th hidden state".format(i))
plt.show()
```
