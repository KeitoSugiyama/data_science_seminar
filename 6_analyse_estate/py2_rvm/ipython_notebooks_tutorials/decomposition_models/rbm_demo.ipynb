# Restricted Boltzmann Machines

RBM is Markov Random Field, which is represented as a undirected bipartite graph that consists of hidden and visible layers. Hidden Layer of RBM is used to model latent space, so in some sense, RBM is trying to do the same thing as PCA and Factor Analysis (i.e. learn latent space representation of data). However, in contrast to  PCA and Factor Analysis, which are directed acyclic graphical models, RBM is a undirected model, which makes inference much more complicated. It was impossible to learn parameters of RBM before Hinton and later Tielman came up with methods to approximate gradient of RBM using Gibbs Sampler ( Hinton - Contrastive Divergence, Tielman - Persistent Contrastive Divergence )

In this notebook, we will show several examples with BernoulliRBM
class implemented in skbayes package.

### What is BernoulliRBM ?

In general Restricted Boltzmann Machines are determined by the form of their hidden and visible units. Hidden neurons are usually binary ( architectures with a different kind of hidden layer did not become popular ), but neurons of the visible layer can have various forms. In BernoulliRBM we assume binary hidden units and visible units that are in range between 0 and 1

### What is difference between Bernoulli RBM implementation in skbayes and sklearn?

 * skbayes implements both CD and PCD (only PCD in sklearn)

 * skbayes implementation has L2 penalty for regularization

 * skbayes uses momentum updates (not simple sgd)

 * skbayes uses adjustable learning rate, which decreases with each iteration

### Example 1: Encoding and Decoding MNIST

This example illustrates the use of BernoulliRBM for dimensionality reduction. At first, we encode original 784-dimensional input data
( mnist image 28-by-28 pixels ) into 200 dimensional latent space, then we reconstruct encoded images.

```python
import numpy as np
from skbayes.decomposition_models import BernoulliRBM
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib
import pandas as pd

# import data ( small subset that you can download from 
# my github: https://github.com/AmazaspShumik/sklearn-bayes/blob/master/ipython_notebooks_tutorials/mixture_models/digits.csv)
# make sure data are in the same directory
data = np.asarray(pd.read_csv('digits.csv'))

rbm = BernoulliRBM(n_components = 200, batch_size = 10, 
                   optimizer = 'cd')

# Note, you can use partial_fit method in case all data
# do not be fit in memory. Code below demonstrates use of partial
# fit.
for i in xrange(20):
    x = data[np.random.randint(0,299,100),:]
    rbm.partial_fit(x)
    
digit_indices = np.asarray([0,101,201])
latent_space = rbm.transform(data[digit_indices,:])
test_data = rbm.reconstruct(latent_space)

fig, axarr = plt.subplots(nrows=3, ncols=2, figsize = (30,30))
print("                Reconstructed  -- vs --  Original")

for i in xrange(len(digit_indices)):
    reconstruction  = (np.reshape(test_data[i,:],(28,28)) > 0.5) * 1
    original        = np.reshape(data[digit_indices[i],:],(28,28))
    im1 = axarr[i,0].imshow(reconstruction, vmin=0, vmax=1,
                            cmap = cm.coolwarm)
    im2 = axarr[i,1].imshow(original, vmin=0, vmax=1,
                            cmap = cm.coolwarm)
    cax,kw = matplotlib.colorbar.make_axes(axarr[i,1])
    plt.colorbar(im1, cax = cax)
    plt.colorbar(im2, cax = cax)
plt.show()
```

As you can see reconstruction quality is quite good.

### Example 2: Encoding and Decoding Olivetti Faces

This example illustrates that in some complicated cases RBM is unable to encode and decode with small reconstruction error. Hence more advanced models are required in this case.

```python
from sklearn.datasets import fetch_olivetti_faces
dataset = fetch_olivetti_faces(shuffle=True, random_state=0)
faces = (dataset.data > 0.5) * 1
```

```python
rbm_olivetti = BernoulliRBM(n_components = 3000, batch_size = 10,
                            optimizer = 'pcd', 
                            l2_penalty = 1e-2)

# incrementally update 
for i in xrange(20):
    x = faces[np.random.randint(0,399,50),:]
    rbm_olivetti.partial_fit(x)
    
original_faces = faces[10:12,:]
transformed_faces = rbm_olivetti.transform(original_faces)
reconstructed_faces = rbm_olivetti.reconstruct(transformed_faces)

fig, axarr = plt.subplots(nrows=2, ncols=2, figsize = (30,24))
print("                Reconstructed  -- vs --  Original")

for i in xrange(len(original_faces)):
    original = np.reshape(original_faces[i,:],(64,64))
    reconstruct = np.reshape(reconstructed_faces[i,:],(64,64))
    im1 = axarr[i,0].imshow(reconstruct, vmin=0, vmax=1, cmap = plt.cm.gray)
    im2 = axarr[i,1].imshow(original, vmin=0, vmax=1, cmap = plt.cm.gray)
    cax,kw = matplotlib.colorbar.make_axes(axarr[i,1])
    plt.colorbar(im1, cax = cax)
    plt.colorbar(im2, cax = cax)
plt.show()
```

### Feature Extraction with RBM

RBM can be used to preprocess data and extract features that can be useful in supervised learning. In this example, we show how latent space representation learned by RBM can enhance classification accuracy on MNIST dataset. We also compare the quality of features extracted by PCA and RBM.
(In this example we use dataset from Kaggle Digit Recognition competition, you can download it here: https://www.kaggle.com/c/digit-recognizer/data )

```python
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import train_test_split
from sklearn.decomposition import PCA
data = pd.read_csv('train.csv')
```

```python
# This tutorial is inspired by tutorial in scikit-learn: 
# http://scikit-learn.org/stable/auto_examples/neural_networks/plot_rbm_logistic_classification.html#sphx-glr-auto-examples-neural-networks-plot-rbm-logistic-classification-py

# make digit features binary
X = (np.asarray(data[data.columns[1:]]) > 0) * 1
Y = np.asarray(data[data.columns[0]])
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                                    test_size=0.5,
                                                    random_state=0)

# preprocess with RBM
rbm_digits = BernoulliRBM(n_components = 350, batch_size = 20,
                          n_iter = 5, verbose = True, 
                          compute_score = True, optimizer = 'pcd')
rbm_digits.fit(X_train)
rbm_features = rbm_digits.transform(X_train)

# preprocess with PCA
# Failed to use n_components = 'mle', had 'math domain error' in log ???
pca_digits = PCA(n_components = 0.8)
pca_features = pca_digits.fit_transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# train Logistic Regression wiht PCA extracted features
lr_pca = LogisticRegression()
lr_pca = lr_pca.fit(pca_features,Y_train)

# train logistic Regression with RBM extracted features
lr_rbm = LogisticRegression()
lr_rbm = lr_rbm.fit(rbm_features,Y_train)

# train logistic Regression with raw pixels
lr     = LogisticRegression()
lr     = lr.fit(X_train,Y_train)
```

```python
# predict and compare perfomance
Y_predict_pca = lr_pca.predict(pca_digits.transform(X_test))
Y_predict_rbm = lr_rbm.predict(rbm_digits.transform(X_test))
Y_predict     = lr.predict(X_test)

print "Logistic Regression using RBM features \n {0}".format(classification_report(Y_test, Y_predict_rbm))
print "Logistic Regression using PCA features \n {0}".format(classification_report(Y_test, Y_predict_pca))
print "Logistic Regression using raw pixels \n {0}".format(classification_report(Y_test, Y_predict))
```
