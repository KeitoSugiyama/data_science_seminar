# Bayesian Bernoulli Mixture Model

This notebook illustrates use of variational Bayesian Bernoulli Mixture for image clustering. We used subset of data used in kaggle digit recognition competition that contains examples of three digits, namely '1','3','6'. 

```python
import numpy as np
import pandas as pd
import pylab
import matplotlib.pyplot as plt
from matplotlib import cm
from skbayes.mixture_models import VBBMM

# load data (you can find data file at: https://github.com/AmazaspShumik/sklearn-bayes/blob/master/ipython_notebooks_tutorials/mixture_models/digits.csv)
digits = np.array(pd.read_csv('digits.csv'))

# initialise bernoulli mixture model with 3 components
n_comoponents = 3;
bmm = VBBMM(n_components = n_comoponents, c = 0.6, d = 0.4, compute_score = True)

# fit model
bmm = bmm.fit(digits)

# expectation of mean / success probabilities of bernoulli components vizualised
fig, axes = plt.subplots(nrows=1, ncols=3, figsize = (40,16))
for i,ax in enumerate(axes):
    im = ax.imshow(np.reshape(bmm.means_[:,i],(28,28)), vmin=0, vmax=1)
    cax,kw = matplotlib.colorbar.make_axes(ax)
    plt.colorbar(im, cax = cax)
plt.show()
```

In figure above you can see means of posterior distribution. Color of each pixel in images above shows probability that pixel will be 'turned on'. If you want to see vector with highest probability of being observed under learned distribution parameters you can use method 'cluster_prototype' as it is illustrated below.

```python
prototypes = bmm.cluster_prototype()
fig, axes = plt.subplots(nrows=1, ncols=3, figsize = (40,16))
for i,ax in enumerate(axes):
    im = ax.imshow(np.reshape(prototypes[:,i],(28,28)), vmin=0, vmax=1, cmap = cm.coolwarm)
    cax,kw = matplotlib.colorbar.make_axes(ax)
    plt.colorbar(im, cax = cax)
plt.show()
```

To find cluster assignments you can use 'predict' method.

```python
bmm.predict(digits)
```

It is possible to monitor convergence of algorithm using lower bound. As you can see from figure below it takes only 10 iterations for variational bayes to converge in this example.

```python

plt.figure(figsize = (10,6))
plt.plot(bmm.scores_[1:])
plt.xlabel('Iterations')
plt.ylabel('value of lower bound')
plt.title('Lower Bound vs Iterations')
plt.show()
```
