# Bayesian Linear Regression

Bayesian Linear Regression (BLR) is very similar to Ridge Regression, however, while Ridge Regression uses cross-validation to find optimal regularization strength, BLR does it analytically. 
There are two different implementations of BLR in skbayes package namely EBLinearRegression and VBLinearRegression. EBLinearRegression uses Empirical Bayes (Type II Maximum Likelihood), while VBLinearRegression uses Variational Bayes. In both cases we put spherical Gaussian prior on weights, however, in Variational Bayes we also add Gamma prior for precision parameter of noise and Gamma hyperprior for the precision parameter of weights. We will not go into further details about differences between Empirical Bayes and Variational Bayes here. You can read more about that in Pattern Recognition and Machine Learning (Bishop 2006) Chapter 3 & 10.

```python
import scipy
import numpy as np
from sklearn.linear_model import LinearRegression, BayesianRidge
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
from sklearn.utils.estimator_checks import check_estimator
from skbayes.linear_models import VBLinearRegression, EBLinearRegression
from sklearn.metrics import mean_squared_error as mse
%matplotlib inline
```

### Both versions of Bayesian Linear Regression pass scikit-learn tests for linear regression model

```python
check_estimator(VBLinearRegression)
check_estimator(EBLinearRegression)
```

### Example 1: Multicollinearity and noisy features

In cases of multicollinearity or a large number of noisy features (some of which can be highly correlated) estimates of coefficients for Ordinary Least Squares become unreliable and have a huge variance. On of the ways to handle this problem is to use Regularized Linear Regression (either L1 or L2), another one is to use Bayesian Linear Regression.
Below we compare estimates of coefficients of OLS EBLinearRegression and VBLinearRegression in a case of 'noisy' features. You can see that estimated coefficients of both versions of Bayesian Regression are much closer to actual weights than estimates from OLS.

```python
    # !!!! This example repeats example from scikit-learn !!!!
    
    # Generate Data Matrix with highly correlated features
    n_samples, n_features              = 100, 100
    np.random.seed(0)
    X  = np.random.random([n_samples,n_features])
        
    # Create weigts with a precision lambda_ of 4.
    lambda_ = 5.
    w = np.zeros(n_features)
    # Only keep 10 weights of interest
    relevant_features = np.random.randint(0, n_features, 10)
    for i in relevant_features:
        w[i] = scipy.stats.norm.rvs(loc=0, scale=1. / np.sqrt(lambda_))
    # Create noise with a precision alpha of 50.
    alpha_ = 50.
    noise = np.random.normal(0,1,n_samples)
    # Create the target
    Y = np.dot(X, w) + noise
    X,x,Y,y   = train_test_split(X,Y, test_size = 0.2, random_state = 1)

    

    # Fit OLS and different versions of bayesian regression
    ols          = LinearRegression()
    ols.fit(X, Y)
    vblr         = VBLinearRegression()
    vblr.fit(X,Y)
    eblr         = EBLinearRegression()
    eblr.fit(X,Y)
    
    # Three subplots sharing both x/y axes
    f, (ax1, ax2) = plt.subplots(2, sharex=True, sharey=False, figsize = (8,10))
    
    ax1.plot(ols.coef_,'b-',label = 'OLS')
    ax1.plot(vblr.coef_,'r-',label = "VBLR")
    ax1.plot(w,'go',label = 'Truth')
    ax1.set_title('VBLR -vs- OLS')
    ax1.legend()
    
    ax2.plot(ols.coef_,'b-',label = 'OLS')
    ax2.plot(eblr.coef_,'r-',label = 'EBLR')
    ax2.plot(w,'go',label = 'Truth')
    ax2.set_title('EBLR -vs- OLS')
    ax2.legend()

    
    # Compute errors on test set 
    olserr   =  mse(y,ols.predict(x))
    eblrerr  =  mse(y,eblr.predict(x))
    vblrerr  =  mse(y,vblr.predict(x))
    # Compute errors on test set 
    olserr   =  mse(y,ols.predict(x))
    eblrerr  =  mse(y,eblr.predict(x))
    vblrerr  =  mse(y,vblr.predict(x))
    print(" ===== MSE on test set ====== ")
    print("   OLS ({0})".format(olserr))
    print("   VBLR({0})".format(vblrerr))
    print("   EBLR({0})".format(eblrerr))
    
```

### Example 2: Overfitting

```python

# number of simulations
n_sims = 50    
# MSEs
mse_vlr, mse_ols, mse_elr = [],[],[]
mse_vlr_train, mse_ols_train = [],[]
mse_elr_train = []
    
    
for i in range(n_sims):
    
        n_samples, n_features              = 100, 120
        X  = np.random.random([n_samples,n_features])
        
        # Create weigts with a precision lambda_ of 4.
        lambda_ = 4.
        w = np.zeros(n_features)
        # Only keep 10 weights of interest
        relevant_features = np.random.randint(0, n_features, 10)
        for i in relevant_features:
            w[i] = scipy.stats.norm.rvs(loc=0, scale=1. / np.sqrt(lambda_))
        # Create noise with a precision alpha of 50.
        alpha_ = 50.
        noise = np.random.normal(0,1,n_samples)
        # Create the target
        Y = np.dot(X, w) + noise

    
        # split into training set and test set
        X,x,Y,y   = train_test_split(X,Y, test_size = 0.2, random_state = 1)
    
        # fit & test models
        vlr       = VBLinearRegression()
        vlr.fit(X,Y)
        y_vlr  = vlr.predict(x)
        mse_vlr.append(mse(y_vlr,y))
        mse_vlr_train.append(mse(vlr.predict(X),Y))
        
        ols         = LinearRegression()
        ols.fit(X,Y)
        y_ols       = ols.predict(x)
        mse_ols.append(mse(y_ols,y))
        mse_ols_train.append(mse(ols.predict(X),Y))
        
        elr       = EBLinearRegression(optimizer = 'em')
        elr.fit(X,Y)
        y_elr = elr.predict(x)
        mse_elr.append(mse(y_elr,y))
        mse_elr_train.append(mse(elr.predict(X),Y))

        
print "\n Average MSE for OLS: training {0}, test {1}".format(np.mean(mse_ols_train),np.mean(mse_ols))        
print "\n Avergae MSE for VBLR: training {0}, test {1}".format(np.mean(mse_vlr_train),np.mean(mse_vlr))
print "\n Average MSE for EBLR: training {0}, test {1} \n".format(np.mean(mse_elr_train),np.mean(mse_elr))
    
# compare mean squared error
plt.figure(2, figsize = (8,6))
bp1 = plt.boxplot((mse_ols,mse_vlr, mse_elr))
plt.title("MSE test set: OLS -vs- VBLR -vs- EBLR -vs- BayesRidge")
plt.ylabel("Mean Squared Error")
plt.show()
```

Note that MSE of Ordinary Least Squares on the test set is much smaller than MSE on training set - a clear sign of overfitting. Results of simulations show no sign of overfitting for EBLinearRegression and VBLinearRegression.

### Example 3: Variance of predicttive distribution

Bayesian Linear Regression can construct predictive distribution for every data point in the test set. This predictive distribution is Gaussian and therefore can be uniquely characterized by mean and standard deviation. It should, however, be noted that in the case of small samples predicted standard deviations could be unreliable.

```python
    # ==== PART IV: Predictive distribution and its properties ==================
    
    # In contrast to BayesianRidge class implemented in scikit-learn our
    # implementation of Bayesian Linear Regression can estimate predictive distribution.
      
    # Small example with predictive distribution
    #np.random.seed(0)
    n,m  = 200,1
    x = np.random.random([n,m])*5 - 10
    br = EBLinearRegression(optimizer = 'em')
    y  = 1 + 2*x[:,0]
    br.fit(x,y)
    y_hat, var = br.predict_dist(x)
    y_lo       = y_hat - 1.96*np.sqrt(var)
    y_hi       = y_hat + 1.96*np.sqrt(var)
    # plotting
    plt.figure(figsize = (12,8))
    plt.plot(x[:,0],y,"bo",label = 'data')
    plt.plot(x[:,0],y_hat,'r+', label = 'model prediction')
    plt.plot(x[:,0],y_lo,'go', label = 'means +- 1.96*std')
    plt.plot(x[:,0],y_hi,'go', label = 'means +- 1.96*std')    
    plt.legend(loc = 2)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title('Predictive distribution')
    plt.show()
```

```python
    # Show how variance of predictive distribution changes with the size of training 
    # data
    
    n_sample   = [20,100,200,1000]
    colors     = ['g-','k-','r-','b-']
    n_test_samples  = 1000
    n_features      = 10
    f, (ax1, ax2) = plt.subplots(2, sharex=True, sharey=False, figsize = (8,13))
    ax1.set_ylim((-0.1,5))
    ax2.set_ylim((-0.1,5))
    for i,n_samples in enumerate(n_sample):
        np.random.seed(10)
        x           = np.random.randn(n_samples,n_features)
        x[0:n_samples,0]  = np.linspace(-100,100,n_samples)
        model  = lambda x: 2 + 2*x[:,0]
        y      = model(x) + np.random.normal(0,1,n_samples)
        eblr2       = EBLinearRegression()
        eblr2.fit(x,y)
        vblr2       = VBLinearRegression()
        vblr2.fit(x,y)
        x_test      = np.random.randn(n_test_samples,n_features)
        x_test[:,0] = np.linspace(-200,200,n_test_samples)
        eblr2_mu, eblr2_var  = eblr2.predict_dist(x_test)
        vblr2_mu, vblr2_var  = vblr2.predict_dist(x_test)
        y_model = model(x_test)
        ax1.plot(x_test[:,0],eblr2_var,colors[i],label = '{0} samples in training'.format(n_samples))
        ax2.plot(x_test[:,0],vblr2_var,colors[i],label = '{0} samples in training'.format(n_samples))
    ax1.set_title('Variance of predictive distribution for EBLR')
    ax2.set_title('Variance of predictive distribution for VBLR')
    ax1.legend(loc = 0)
    ax2.legend(loc = 0)
    plt.show()
```
